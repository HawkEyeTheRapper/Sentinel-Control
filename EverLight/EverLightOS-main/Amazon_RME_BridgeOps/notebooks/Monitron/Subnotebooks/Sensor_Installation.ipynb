# Sensors — Installation & Proper Placement (Ghost-Alert Kill Kit)

> If a TXV bulb is mounted wrong, the whole system lies to you. Same here: a poorly mounted Monitron sensor will invent “faults” (or hide real ones). This page is your **do-it-right-once** guide.

---

## Core Principles (do these every time)

**1) Mount on the load path (not on sheet metal).**  
Place the sensor on the **bearing housing / gearbox case nearest the loaded zone**, not on guards, covers, or baseplates.

**2) Prep the surface.**  
- Degrease → scrape/abrade to bare metal → wipe with IPA → dry.  
- Flat, rigid contact patch ≈ coin-size. No paint, no rust, no gasket line.

**3) Rigid > convenient.**  
Flexible/floating parts (thin guards, long brackets) resonate and inject noise. If you can wiggle it by hand, don’t mount there.

**4) Orientation & axes.**  
Tri-axial sensors are forgiving, but **radial** (perpendicular to shaft) is most fault-revealing; **axial** helps spot misalignment. If you must choose one face, bias radial.

**5) Thermal sanity.**  
Keep clear of hot air streams and sun-baked panels. You want **component surface temperature**, not ambient wash.

**6) Label + photo on install.**  
Name the point consistently (e.g., `GBX_OUT_RADIAL`), snap a picture, and attach it to the asset record. Future you will thank you.

---

## Component Cheat-Sheet (quick picks)

**Motors (coupled):**  
- Primary: **DE (drive-end) bearing housing**, radial; Secondary: **NDE**, axial.  
- Avoid: fan covers, endbells with thin shrouds.

**Gearboxes:**  
- Input and output bearing housings (both sides).  
- Avoid: top cover plates and long unsupported lids.

**Pumps/Fans/Blowers:**  
- Pump: bearing pedestal near wet end; Fan: near drive-side pillow block.  
- Avoid: volute casings/thin guards.

**Conveyors (belt/roller):**  
- Motor DE, coupling, gearbox in/out; **drive roller L/R**, **tension roller L/R**, **end roller L/R**.  
- Avoid: sheet-metal stringers and bolt-on guards.

---

## Commissioning Checklist (10 minutes)

**A. Before you stick it**  
- [ ] Surface prepped to bare metal, dry, rigid.  
- [ ] Location is on the **component housing** (not the structure).  
- [ ] BLE link to gateway is likely OK (line-of-sight if possible).

**B. Stick & seat**  
- [ ] Press firmly 30–60 s per adhesive guidance (or apply epoxy per spec).  
- [ ] Note the **point name** you’ll use in the asset map.

**C. Link test (BLE sanity)**  
- [ ] App sees the sensor; RSSI “good” (rule-of-thumb: better than ~-70 dBm; at ~-85 dBm, add a gateway or move it).  
- [ ] First reading arrives (or force a check-in if supported).

**D. Coupling test (optional but ideal)**  
- [ ] Light tap on housing near sensor with plastic handle: you should see a short, clean response—not a mushy smear. Mushy → weak mounting or flimsy structure.

**E. Baseline stamp**  
- [ ] Record **state-matched** baseline (RPM/load/temp).  
- [ ] Attach photo + note to asset: “Sensor on DE bearing housing, radial.”

---

## Ghost-Alert Killers (what bites most often)

- ❌ **Painted/rusty/oily surface** → fake broad-band noise, poor repeatability.  
- ❌ **On a guard/cover** → resonant harmonics unrelated to the bearing.  
- ❌ **Across a gasket line** → decoupled signal, temperature drift.  
- ❌ **Too far from the fault** (e.g., baseplate) → low SNR, misses early defects.  
- ❌ **Hot air stream/sun** → temp trends that aren’t machine heat.  
- ❌ **BLE marginal** → intermittent data → “choppy” trends that look like faults.

**Fix order:** relocate to rigid housing → re-prep → re-mount → re-baseline → verify BLE.

---

## Acceptance Criteria (you can trust this install if…)

- **Geometry:** on housing, near load zone; not on covers/frames.  
- **Signal:** repeatable baseline (RMS variance small on like-for-like runs).  
- **Trend:** no mysterious spikes after 24–48 h of steady operation.  
- **Comm:** stable check-ins; no repeated dropouts.

# Sensors — Installation & Proper Placement (Ghost-Alert Kill Kit)

> A bad mount lies to you. Treat this like a TXV bulb: **clean surface, correct spot, firm contact**—or your data will mislead you.

---

## Where to place (principles)
- **Mount on the component housing** you care about (bearing housing / gearbox case), **not** on guards or thin sheet metal.  
- **Minimize the transmission path**: shorter, stiffer paths carry true vibration; rubber/plastic and multiple joints attenuate & reflect signal.  
- Expect useful energy to fade by roughly **75–90 cm (30–36")** from the source—get close.  
- Choose points that won’t **oscillate at natural frequencies** (avoid floppy covers).  
- For tri-axial sensors, prefer **radial** to the shaft; add **axial** on misalignment-prone points.

## How to mount (mechanics)
- Prep: **degrease → abrade to bare metal → IPA wipe → dry**. Flat, rigid patch ≈ coin-size.  
- **Seat ≥ 1/3 of the sensor base** solidly on the target area; center the target under the sensor.  
- **Adhesive**: cyanoacrylate-type per surface condition.  
  - Flat (<0.1 mm roughness): **LOCTITE 454**  
  - Gaps/rough (~≤5 mm): **LOCTITE 3090** or **LOCTITE 4070**  
- Press/hold per adhesive spec; let it cure properly before baseline.

## Typical positions (quick picks)
- **Motor (coupled):** Drive-end bearing (radial); non-drive-end (axial as secondary).  
- **Gearbox:** Input & output bearing housings.  
- **Pumps/Fans/Blowers:** Bearing pedestal near wet end / drive-side pillow block.  
- **Conveyors:** Motor DE, coupling, gearbox in/out; drive, tension, and end rollers (L/R).

## Commission & baseline
- Pair via **NFC** in the mobile app; pick **Position Type** (bearing, motor, pump, etc.) and **Class** (ISO 20816 derived)—note: **position type can’t be changed after**.  
- **Baseline only in a healthy state** so the model learns “normal.”  
- Use the app/web to monitor **warning/alert** states (ML + ISO thresholds) and resolve with a clear failure mode/cause entry.

## Gateway sanity (so readings actually arrive)
- Mount gateways **higher than sensors** and within **~20–30 m** of them; commission via the mobile app (Wi-Fi or Ethernet flow).  

## Ghost-alert killers (most common mistakes)
- Painted/oily/rusty surface → noisy, inconsistent spectra.  
- Mounted on **guards/covers** → resonant harmonics unrelated to the bearing.  
- Across a **gasket seam** → thermal/vibration decoupling.  
- Too far from the fault → low SNR; misses early defects.  
- Ambient hot-air/sun on the sensor → fake temperature trends.

## Quick acceptance check (trust this install if…)
- **Geometry:** on rigid housing, near the load path (not structure/guards).  
- **Signal:** stable baseline on like-for-like runs; no random spikes over 24–48h.  
- **Comms:** gateway commissioned; sensor check-ins steady in the app.
