```python
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell, new_code_cell
from pathlib import Path

ROOT = Path("/mnt/dev/omniversal_dev/EyeOfAether/AetherIntelligence/The_Black_Swan_Accords")

def lineage_roles_for(book_meta):
    roles = []
    pcs = book_meta.get("primary_characters", [])
    if "Selene" in pcs:
        roles.append("Selene — Last Sentinel (Will)")
    if "Halana" in pcs:
        roles.append("Halana — Last Sunspeaker (Light)")
    if "Alexia" in pcs:
        roles.append("Alexia — Celestial Heir (Ascension)")
    return roles


def make_series_master_notebook(arc_id, series_meta):
    series_title = series_meta["series_title"]
    tagline = series_meta.get("tagline", "")
    themes = series_meta.get("themes", [])
    books = series_meta.get("books", {})

    header_md = f"# {series_title} ({arc_id})\n\n{tagline}"
    themes_md = (
        "### Themes\n" + ", ".join(f"`{t}`" for t in themes)
        if themes else "### Themes\n_(to be defined)_"
    )

    lines = ["### Books"]
    for book_id, book_meta in sorted(books.items(), key=lambda kv: kv[1].get("sequence", 0)):
        rel = f"../{book_id}/Master.ipynb"
        lines.append(f"- **{book_meta['title']}** (`{book_id}`) — [open]({rel})")
    books_md = "\n".join(lines)

    nb = new_notebook(
        cells=[
            new_markdown_cell(header_md),
            new_markdown_cell(themes_md),
            new_markdown_cell(books_md),
        ]
    )
    return nb


def make_book_master_notebook(arc_id, series_meta, book_id, book_meta):
    title = book_meta["title"]
    series_title = series_meta["series_title"]
    seq = book_meta.get("sequence")
    tags = book_meta.get("tags", [])
    primary_characters = book_meta.get("primary_characters", [])
    links = book_meta.get("links", {})
    lineage_roles = lineage_roles_for(book_meta)

    header_md = (
        f"# {title}\n\n"
        f"**Series:** {series_title} ({arc_id})  \\\n"
        f"**Book ID:** `{book_id}`  \\\n"
        f"**Sequence:** {seq}"
    )

    tags_md = (
        "### Tags\n" + ", ".join(f"`{t}`" for t in tags)
        if tags else "### Tags\n_(none yet)_"
    )

    chars_md = (
        "### Primary Characters\n" + ", ".join(primary_characters)
        if primary_characters else "### Primary Characters\n_(to be defined)_"
    )

    if lineage_roles:
        lineage_md = "### Lineage Thread\n" + "\n".join(f"- {r}" for r in lineage_roles)
    else:
        lineage_md = "### Lineage Thread\n- _(no direct lineage anchor in this volume)_"

    nav_lines = []
    nav_lines.append("### Navigation")
    nav_lines.append("- [⬆ Back to Series Master](../00_Series_Master/Master.ipynb)")
    if links:
        nav_lines.append("")
        nav_lines.append("**Linked Volumes Across Arcs:**")
        for key, target in links.items():
            if not target:
                continue
            rel_path = f"../../{target}/Master.ipynb"
            label = key.capitalize()
            nav_lines.append(f"- **{label}** → [{target}]({rel_path})")
    nav_md = "\n".join(nav_lines)

    scratch_md = """---
### Working Notes

Use this space for:
- scene lists
- structural beats
- questions for future revision
- cross-references to VALOR / The Unveiling / Nightfall

You can always regenerate this notebook from `FormingTheAccords.ipynb` if the structure changes.
"""

    nb = new_notebook(
        cells=[
            new_markdown_cell(header_md),
            new_markdown_cell(tags_md),
            new_markdown_cell(chars_md),
            new_markdown_cell(lineage_md),
            new_markdown_cell(nav_md),
            new_markdown_cell(scratch_md),
        ]
    )
    return nb


# === Driver ===
for arc_id, series_meta in black_swan_meta.items():
    arc_root = ROOT / arc_id
    if not arc_root.exists():
        continue

    series_master_dir = arc_root / "00_Series_Master"
    series_master_path = series_master_dir / "Master.ipynb"
    series_master_dir.mkdir(parents=True, exist_ok=True)

    if not series_master_path.exists():
        nb = make_series_master_notebook(arc_id, series_meta)
        nbformat.write(nb, str(series_master_path))
        print(f"✅ Wrote series master {series_master_path}")
    else:
        print(f"⚪ Series master exists, skipping: {series_master_path}")

    for book_id, book_meta in series_meta["books"].items():
        book_path = arc_root / book_id
        if not book_path.exists():
            continue

        master_path = book_path / "Master.ipynb"
        if master_path.exists():
            print(f"⚪ Book master exists, skipping: {master_path}")
            continue

        nb = make_book_master_notebook(arc_id, series_meta, book_id, book_meta)
        nbformat.write(nb, str(master_path))
        print(f"✅ Wrote book master {master_path}")

print("✨ Master notebooks generation complete.")
```

```python
import nbformat
from nbformat.v4 import new_notebook, new_markdown_cell, new_code_cell
from pathlib import Path

# Same ROOT as above – keep this in sync with your earlier cells
ROOT = Path("/mnt/dev/omniversal_dev/EyeOfAether/AetherIntelligence/The_Black_Swan_Accords")

def lineage_roles_for(book_meta):
    """Return a list of lineage strings based on primary characters."""
    roles = []
    pcs = book_meta.get("primary_characters", [])
    if "Selene" in pcs:
        roles.append("Selene — Last Sentinel (Will)")
    if "Halana" in pcs:
        roles.append("Halana — Last Sunspeaker (Light)")
    if "Alexia" in pcs:
        roles.append("Alexia — Celestial Heir (Ascension)")
    return roles


def make_series_master_notebook(arc_id, series_meta):
    """Create the 00_Series_Master/Master.ipynb notebook for an arc."""
    series_title = series_meta["series_title"]
    tagline = series_meta.get("tagline", "")
    themes = series_meta.get("themes", [])
    books = series_meta.get("books", {})

    header_md = f"# {series_title} ({arc_id})\n\n{tagline}"
    themes_md = (
        "### Themes\n" + ", ".join(f"`{t}`" for t in themes)
        if themes else "### Themes\n_(to be defined)_"
    )

    # Book list with links to each book's Master notebook
    lines = ["### Books"]
    for book_id, book_meta in sorted(books.items(), key=lambda kv: kv[1].get("sequence", 0)):
        rel = f"../{book_id}/Master.ipynb"
        lines.append(f"- **{book_meta['title']}** (`{book_id}`) — [open]({rel})")
    books_md = "\n".join(lines)

    nb = new_notebook(
        cells=[
            new_markdown_cell(header_md),
            new_markdown_cell(themes_md),
            new_markdown_cell(books_md),
        ]
    )
    return nb


def make_book_master_notebook(arc_id, series_meta, book_id, book_meta):
    """Create the per-book Master.ipynb notebook."""
    title = book_meta["title"]
    series_title = series_meta["series_title"]
    seq = book_meta.get("sequence")
    tags = book_meta.get("tags", [])
    primary_characters = book_meta.get("primary_characters", [])
    links = book_meta.get("links", {})
    lineage_roles = lineage_roles_for(book_meta)

    header_md = (
        f"# {title}\n\n"
        f"**Series:** {series_title} ({arc_id})  \\\n"
        f"**Book ID:** `{book_id}`  \\\n"
        f"**Sequence:** {seq}"
    )

    tags_md = (
        "### Tags\n" + ", ".join(f"`{t}`" for t in tags)
        if tags else "### Tags\n_(none yet)_"
    )

    chars_md = (
        "### Primary Characters\n" + ", ".join(primary_characters)
        if primary_characters else "### Primary Characters\n_(to be defined)_"
    )

    if lineage_roles:
        lineage_md = "### Lineage Thread\n" + "\n".join(f"- {r}" for r in lineage_roles)
    else:
        lineage_md = "### Lineage Thread\n- _(no direct lineage anchor in this volume)_"

    # Cross-arc navigation using the `links` map in your metadata
    nav_lines = []
    nav_lines.append("### Navigation")
    nav_lines.append("- [⬆ Back to Series Master](../00_Series_Master/Master.ipynb)")
    if links:
        nav_lines.append("")
        nav_lines.append("**Linked Volumes Across Arcs:**")
        for key, target in links.items():
            if not target:
                continue
            # Example target: "02_The_Unveiling/01_First_Light"
            rel_path = f"../../{target}/Master.ipynb"
            label = key.capitalize()  # "valor", "unveiling", "nightfall"
            nav_lines.append(f"- **{label}** → [{target}]({rel_path})")
    nav_md = "\n".join(nav_lines)

    scratch_md = """---
### Working Notes

Use this space for:
- scene lists
- structural beats
- questions for future revision
- cross-references to VALOR / The Unveiling / Nightfall

You can always regenerate this notebook from `FormingTheAccords.ipynb` if the structure changes.
"""

    nb = new_notebook(
        cells=[
            new_markdown_cell(header_md),
            new_markdown_cell(tags_md),
            new_markdown_cell(chars_md),
            new_markdown_cell(lineage_md),
            new_markdown_cell(nav_md),
            new_markdown_cell(scratch_md),
        ]
    )
    return nb


# === Driver: generate all series masters + book masters ===
for arc_id, series_meta in black_swan_meta.items():
    arc_root = ROOT / arc_id
    if not arc_root.exists():
        continue

    # 1) Series master notebook
    series_master_dir = arc_root / "00_Series_Master"
    series_master_path = series_master_dir / "Master.ipynb"
    series_master_dir.mkdir(parents=True, exist_ok=True)

    if not series_master_path.exists():
        nb = make_series_master_notebook(arc_id, series_meta)
        nbformat.write(nb, series_master_path)
        print(f"✅ Wrote series master {series_master_path}")
    else:
        print(f"⚪ Series master exists, skipping: {series_master_path}")

    # 2) Per-book master notebooks
    for book_id, book_meta in series_meta["books"].items():
        book_path = arc_root / book_id
        if not book_path.exists():
            continue

        master_path = book_path / "Master.ipynb"
        if master_path.exists():
            print(f"⚪ Book master exists, skipping: {master_path}")
            continue

        nb = make_book_master_notebook(arc_id, series_meta, book_id, book_meta)
        nbformat.write(nb, master_path)
        print(f"✅ Wrote book master {master_path}")

print("✨ Master notebooks generation complete.")
```

```python
import json
from pathlib import Path

ROOT = Path("/mnt/dev/omniversal_dev/EyeOfAether/AetherIntelligence/The_Black_Swan_Accords")

for arc_id, arc_cfg in black_swan_meta.items():
    for book_id, book_meta in arc_cfg["books"].items():
        book_path = ROOT / arc_id / book_id
        if not book_path.exists():
            continue  # skip if folder hasn't been created yet

        out = {
            "series_id": arc_cfg["series_id"],
            "series_title": arc_cfg["series_title"],
            "book_id": book_id,
            "book_title": book_meta["title"],
            "sequence": book_meta.get("sequence"),
            "tags": book_meta.get("tags", []),
            "primary_characters": book_meta.get("primary_characters", []),
            "links": book_meta.get("links", {}),
        }

        with open(book_path / "metadata.json", "w", encoding="utf-8") as f:
            json.dump(out, f, ensure_ascii=False, indent=2)

print("✅ metadata.json files written for all mapped books.")
```
