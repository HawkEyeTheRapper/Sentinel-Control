````python
# Boot: EverLight Aether Indexer (Tinker-ready)
import os, json, glob, yaml, hashlib, time
from pathlib import Path

REPO = Path("..")  # Go up one level to repo root
MANIFEST = yaml.safe_load((REPO / "aether_manifest.yaml").read_text())
LINKS = json.loads((REPO / "aether_links.yml").read_text())

def gather_files(patterns):
    files = []
    for pat in patterns:
        files.extend(glob.glob(pat, recursive=True))
    # de-dupe and ignore binary blobs by extension
    exts_ok = {".md",".txt",".ipynb",".yaml",".yml",".json"}
    files = [f for f in sorted(set(files)) if Path(f).suffix.lower() in exts_ok]
    return files

def make_doc(file_path, arc, tags):
    p = Path(file_path)
    text = p.read_text(errors="ignore")
    return {
        "id": hashlib.sha1((str(p)+str(os.path.getmtime(p))).encode()).hexdigest(),
        "path": str(p),
        "arc": arc,
        "tags": tags,
        "content": text
    }

docs_by_index = {}
for idx in MANIFEST["vector_indices"]:
    files = gather_files(idx["include"])
    docs = [ make_doc(f, idx["scope"], idx["metadata"]["tags"]) for f in files ]
    docs_by_index[idx["name"]] = docs
    print(f"Prepared {len(docs):>4} docs for index {idx['name']}")

# 🔧 Tinker client placeholder — swap in the real calls
class TinkerClient:
    def upsert(self, index_name, docs, namespace=None):
        # TODO: replace with tinker.index.upsert(...)
        print(f"[Tinker] Upserting {len(docs)} docs to {index_name} ns={namespace}")

tinker = TinkerClient()
ns = MANIFEST["tinker_pipelines"]["namespace_prefix"]

for index_name, docs in docs_by_index.items():
    tinker.upsert(index_name, docs, namespace=f"{ns}:{index_name}")
````

````python
# Character analysis across series
def load_characters(series_path):
    """Load character files from a series directory"""
    characters = {}
    char_path = series_path / 'characters'
    if char_path.exists():
        for char_file in char_path.glob('*.txt'):
            with open(char_file, 'r', encoding='utf-8') as f:
                characters[char_file.stem] = f.read()
    return characters

# Load all characters
valor_chars = load_characters(valor_path)
nightfall_chars = load_characters(nightfall_path)

print(f"VALOR Characters: {list(valor_chars.keys())}")
print(f"Nightfall Characters: {list(nightfall_chars.keys())}")
````
